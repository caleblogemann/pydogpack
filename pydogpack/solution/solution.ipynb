```python
from pydogpack.solution import solution
from pydogpack.basis import basis
from pydogpack.mesh import mesh
from pydogpack.utils import x_functions
```

# DG Solution Object

## One Dimension

```python
# Initial Condition
ic = x_functions.Sine()
mesh_ = mesh.Mesh1DUniform(0.0, 1.0, 10)
basis_ = basis.LegendreBasis1D(3)
dg_solution = basis_.project(ic, mesh_)
```

```python
dg_solution.create_plot();
```

## Two Dimensions

```python
ic = x_functions.Sine2D()
```

### Cartesian Mesh

```python
mesh_ = mesh.Mesh2DCartesian(-1.0, 1.0, -1.0, 1.0, 10, 10)
basis_ = basis.LegendreBasis2DCartesian(2)
dg_solution = basis_.project(ic, mesh_)
dg_solution.create_plot();
```

### Unstructured Mesh

```python
mesh_ = mesh.Mesh2DMeshGenDogPack("../mesh/square_sample_mesh")
basis_ = basis.ModalBasis2DTriangle(3)
dg_solution = basis_.project(ic, mesh_)
dg_solution.create_plot();
```

```python
mesh_ = mesh.Mesh2DTriangulatedRectangle(0.0, 1.0, 0.0, 1.0, 20, 20)
basis_ = basis.ModalBasis2DTriangle(4)
dg_solution = basis_.project(ic, mesh_)
dg_solution.create_plot();
```

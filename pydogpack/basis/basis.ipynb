```python
from pydogpack.basis import basis
from pydogpack.mesh import mesh
from pydogpack.utils import x_functions
import numpy as np
```

# Basis

## 1 Dimensional Bases

### Nodal Basis

```python
nodes = np.linspace(-1, 1, 5)
nodal_basis_1d = basis.NodalBasis1D(nodes)
nodal_basis_1d.create_plot();
```

```python
ic = x_functions.Sine()
mesh_ = mesh.Mesh1DUniform(0.0, 1.0, 10)
dg_solution = nodal_basis_1d.project(ic, mesh_)
dg_solution.create_plot();
```

## 2 Dimensional Bases

### Legendre Modal Basis Rectangular

```python
legendre_basis_2d_cartesian = basis.LegendreBasis2DCartesian(3)
for i_basis_cpt in range(legendre_basis_2d_cartesian.num_basis_cpts):
    legendre_basis_2d_cartesian.create_plot(i_basis_cpt);
```

```python
ic = x_functions.Sine2D()
mesh_ = mesh.Mesh2DCartesian(0.0, 1.0, 0.0, 1.0, 10, 10)
dg_solution = legendre_basis_2d_cartesian.project(ic, mesh_)
dg_solution.create_plot();
```

### Modal Basis Triangle

```python
modal_basis_2d_triangle = basis.ModalBasis2DTriangle(3)
for i_basis_cpt in range(modal_basis_2d_triangle.num_basis_cpts):
    modal_basis_2d_triangle.create_plot(i_basis_cpt);
```

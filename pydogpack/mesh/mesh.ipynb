```python
from pydogpack.mesh import mesh
import numpy as np
```

# Mesh

## One Dimensional Meshes

### Uniform Mesh

```python
mesh_1d_uniform = mesh.Mesh1DUniform(0.0, 1.0, 10)
mesh_1d_uniform.create_plot();
```

### Custom Mesh

```python
vertices = np.array([[-1.0], [-0.5], [-0.2], [0.0], [0.3], [0.7], [1.0]])
mesh_1d = mesh.Mesh1D(vertices)
mesh_1d.create_plot();
```

## Two Dimensional Meshes

### Cartesian Rectangle

```python
mesh_2d_cartesian = mesh.Mesh2DCartesian(0.0, 1.0, 0.0, 1.0, 10, 10)
mesh_2d_cartesian.create_plot();
```

### Triangulated Rectangle

```python
mesh_2d_triangulated_rectangle = mesh.Mesh2DTriangulatedRectangle(0.0, 1.0, 0.0, 1.0, 10, 10)
mesh_2d_triangulated_rectangle.create_plot();
```
